```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv('Dataset_Thuy.csv')
data_copy= data.copy()

# Enlever les variables qui ne contribuent pas à la prédictiono tel que les identifiant . 
# Invalidity et Infringment sont des variables qui vont nous permettre de créer notre variable cible: 

data_copy=data_copy.drop(columns=['appln_id','pub_nbr','patent','Infringment','Invalidity'])

#création de la variable de prédiction 'litige'
variables=data[['Infringment','Invalidity']]



prediction=np.zeros(variables.shape[0])
for i in range (variables.shape[0]):
    if variables['Infringment'][i]==1:
        prediction[i]=1
    if variables['Invalidity'][i]==1:
        prediction[i]=1


#ajt de la nvl variable dans la dataframe 

data_copy['prediction']=prediction

data_copy.head().T


```

```python
print(data_copy.isna().sum())
```

Il y a trop de données manquantes pour la variables generality et Year of litigation , enlevons les

```python
data_copy=data_copy.drop(columns=['Year_Litigation'])
```

```python
data_copy['originality'] = data_copy['originality'].fillna(data_copy['originality'].mean())
data_copy['generality'] = data_copy['generality'].fillna(data_copy['generality'].mean())
```

```python
data_copy.dropna(subset=['claims'],inplace=True)
data_copy.dropna(subset=['quality_index_4'],inplace=True) 

print(data_copy.isna().sum())
```
